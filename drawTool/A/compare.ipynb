```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

csv_file = 'compare_result.csv'
df = pd.read_csv(csv_file)

gpt4o = df[['Q2', 'Q3', 'Q4']].mean()
llama2gpt4o = df[['Q5', 'Q6', 'Q7']].mean()

categories = ['Code Formatting', 'Code Logic', 'Code Structure']

results_df = pd.DataFrame({
    # 'Category': categories,
    'gpt-4o': gpt4o.values,
    'llama2gpt4o': llama2gpt4o.values
}, index=categories)

results_df.to_csv("compared_data.csv", index=False)

x = np.arange(len(categories))
width = 0.35 

fig, ax = plt.subplots()
plt.ylim(0, 1.2)
rects1 = ax.bar(x - width/2, gpt4o, width, label='gpt-4o', color='blue')
rects2 = ax.bar(x + width/2, llama2gpt4o, width, label='llama2gpt4o', color='orange')

ax.set_ylabel('Scores')
ax.set_title('Comparison of GPT-4o and LLaMA2GPT4o by Category')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

def add_labels(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}', xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

add_labels(rects1)
add_labels(rects2)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

csv_file = 'compare_result.csv'
df = pd.read_csv(csv_file)

gpt4o_counts = df[['Q2', 'Q3', 'Q4']].sum()
llama2gpt4o_counts = df[['Q5', 'Q6', 'Q7']].sum()

categories = ['Code Formatting', 'Code Logic', 'Code Structure']

x = np.arange(len(categories))
width = 0.35

fig, ax = plt.subplots()
plt.ylim(0, 330)
rects1 = ax.bar(x - width/2, gpt4o_counts, width, label='gpt-4o', color='blue')
rects2 = ax.bar(x + width/2, llama2gpt4o_counts, width, label='llama2gpt4o', color='orange')

ax.set_ylabel('Count of 1s')
ax.set_title('Comparison of GPT-4o and LLaMA2GPT4o by Category')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

def add_labels(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{int(height)}', xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

add_labels(rects1)
add_labels(rects2)

# 显示图表
plt.tight_layout()
plt.show()
```
